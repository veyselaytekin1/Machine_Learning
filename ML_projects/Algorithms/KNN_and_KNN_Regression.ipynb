### gercek veri ile bir uygulama

```python
from sklearn.datasets import load_breast_cancer

#kanser veri seti
```

```python
cancer = load_breast_cancer()
cancer
```

```python
cancer.keys()
```

```python
cancer['DESCR']
#veri seti ile ilgili bilgiler
```

```python
X_train, X_test, y_train,y_test = train_test_split(cancer.data, 
                                                   cancer.target,
                                                   stratify=cancer.target,
                                                   random_state=66)
```

```python
X_train
```

```python
train_dogruluk = []
test_dogruluk = []

#biz hangi n_neighbors degerinin daha basarili oldugunu bilmiyoruz
#bunu sirayla deneyecegiz
```

```python
komsuluk_sayisi = range(1,11)
```

```python
 for n_komsuluk in komsuluk_sayisi:
        snf=KNeighborsClassifier(n_neighbors=n_komsuluk)
        snf.fit(X_train,y_train)
        train_dogruluk.append(snf.score(X_train,y_train))
        test_dogruluk.append(snf.score(X_test,y_test))

print(train_dogruluk)  
print()
print(test_dogruluk)        
```

```python
import matplotlib.pyplot as plt
plt.plot(komsuluk_sayisi,train_dogruluk, label ='Train Dogruluk')
plt.plot(komsuluk_sayisi,test_dogruluk, label ='Test Dogruluk')
plt.ylabel= 'Dogruluk'
plt.xlabel= 'n_komsuluk'
plt.legend()
#bunlari daha iyi görmek icin grafik cizelim

#bu veri setinde 6 komsuluga göre yapilan daha basarilidir
```
