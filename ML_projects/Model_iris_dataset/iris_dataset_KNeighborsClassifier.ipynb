## veri setini bölme

```python
from sklearn.model_selection import train_test_split
```

```python
X = iris['data']
y = iris['target']

#burda veri setinde hangilerinin X ve y oldugunu belirledik
#X_train, X_test, y_train, y_test = train_test_split(iris['data'],iris['target]', train_size=0.80, random_state=0 )
#olarakta yazabilirdik 
```

## Model kurma

```python
#sklearnde kullanabilecegimiz bircok siniflandirma vardir
#burda anlasilmasi en kolay olan K en yakin komsu siniflandirmasini kullanalim
#model egitme yalniz egitim verisi üzerinden yapilir
#yeni bir veri noktasi icin algoritma bu noktaya en yakin egitim noktasindaki veriyi bulur
#bu egitim verisi noktasinin etiketi yeni veri noktasina atanir
#KNN komsu algoritmasini kullanmak icin KNeighborsClasifier i import edelim
```

```python
from sklearn.neighbors import KNeighborsClassifier 

knn = KNeighborsClassifier(n_neighbors=1)


#bu sinifi secelim
#daha sonra knn adinda bir degisken alalim
#ve bu degisken icin KNeighborsClasifier  kullanip, icerisine n_neighbors parametresi icin 1 yazalim
```

```python
knn.fit(X_train, y_train)


#modeli kurmak icin knn nesnesinin fit methodunu cagiriyoruz
#knn fit icerisinde  X_train ve y_train yazip bu komutu calistiriyorum
#bu methot knn nesnesinin kendisini döndürdü
#bu ifadeler olustuulan modeldeki parametreleri gösterir
#bu parametrelerin cogu ön tanimlidir
#yerine göre bu parametreleri modifiye edebiliriz
```

```python
yeni_cicek = np.array([[5,2.9,1,0.2]])

#bu hangi sinifta oldugunu ögrenmek istedigim yeni cicegin ölcüm degerleri
```

```python
yeni_cicek_sinifinin_tahmini = knn.predict(yeni_cicek)
yeni_cicek_sinifinin_tahmini
#yani setosa dedi,ama yazimi güzel olsun diye tirendasinkini yazayim asagiya
```

```python
print('tahmin sinifi',yeni_cicek_sinifinin_tahmini) #yeni_cicek_sinifinin_tahmini demek 0 demek
print('tahmin türü',iris['target_names'][yeni_cicek_sinifinin_tahmini])
```

## Modelin Performansi

```python
#kurdugumuz modelle yeni bir iris ciceginin türünü tahmin edebiliyoruz, peki modele ne kadar güveniyoruz
```

```python
knn_pred = knn.predict(X_test)

#simdi bu modeli kullanarak yeni veriler icin tahminler olusturalim


knn_pred

# 0=setosa
# 1=versicolor
# 2=virginica
```

```python
y_test
#buda gercek degerler. ve mükemmel bir sekilde tahmin etti zaten asagidada performansi 1 cikti, 
#%100 dogrulukla dogru tahmin etti
```

```python
np.mean(knn_pred==y_test)

#tirendaz yukaridaki yaptigi tahminleri gercek degrlriyle karsilastiralim dedi,
#bu veri icin bu sekilde kolayca görebiliriz ama büyük veriler icin baska yol lazim
#bunu numpy daki mean methodunu kullnarak yapabiliriz
#np.mean yazip icerisine knn_pred==y_test yazalim 
#ve tahmin edilen bu türlerin gercek türle nekadar esit oldugunun ortalamasini aliyoruz
#bu komutu calistirdigim zaman bu modelim dogrulugu 1 cikti
```

```python
knn.score(X_test, y_test).round(3) 

#buda ikinci bir yol, bir üsteki tirendaz yapti
```
